# Feature engineering

```python
from sklearn.model_selection import train_test_split
x_train ,x_test, y_train,y_test = train_test_split(df.drop(columns=['target']),df['target'],test_size=0.2,random_state=42)
```

```python
x_train.shape , x_test.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
```

```python
x_train.head(2)
```

```python
trf1 = ColumnTransformer([
    ("ohe_trf",OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore'),[1,2,5,6,8,10,11])
], remainder='passthrough')
```

```python
trf1
```

```python
trf2 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
trf2
```

```python
from sklearn.linear_model import LogisticRegression
trf3 = LogisticRegression()
```

```python
trf3
```

***Creating pipeline***

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(trf1,trf2,trf3)
```

### Train the model and predict the model on test data

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,x_train,y_train,cv=5,scoring='accuracy').mean()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of Logistic Regression",accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred),columns=list(range(0,2)))
```

```python
result = pd.DataFrame()
result['Actual Label'] = y_test
result['Logistic Regression Prediction'] = y_pred
```
